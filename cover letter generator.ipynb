```python
import os
from langchain_openai import ChatOpenAI
from langchain_community.document_loaders import PyMuPDFLoader

from dotenv import load_dotenv
load_dotenv()
```

```python
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")

llm=ChatOpenAI(
    model_name="gpt-3.5-turbo"
)
llm.invoke("who is lincon. answer in 1 line")
```

```python
loader=PyMuPDFLoader("/Users/mojo/Generative AI/Projects/Resume-Genie/Manoj Kumar-Gen AI.pdf")
resume = loader.load()
# resume[0].page_content

# clean view
context = "\n\n".join([d.page_content for d in resume])
```

```python
job_description="""
We are seeking a talented AI Engineer with at least 2 years of experience building AI-driven solutions. The ideal candidate will design, develop, and deploy scalable machine learning and generative AI systems to solve real-world problems.

Key Responsibilities:

Build and deploy AI/ML models and pipelines
Develop and optimize LLM-based applications (RAG, prompt engineering, etc.)
Work with large datasets for training and evaluation
Collaborate with cross-functional teams to integrate AI solutions
Monitor model performance and improve accuracy over time

Required Skills:

2+ years of experience building AI or ML solutions
Strong proficiency in Python and ML frameworks (TensorFlow, PyTorch)
Experience with LLMs, NLP, or Generative AI
Knowledge of cloud platforms (AWS, Azure, or GCP)
Familiarity with data pipelines and APIs"""
```

```python
template = """
You are an expert career coach and professional cover letter writer.

Your task is to generate a highly personalized and compelling cover letter based on the candidate's resume and the job description.

Candidate Resume:
{context}

Job Description:
{job_description}

Instructions:
1. Tailor the cover letter specifically to the job role and company.
2. Highlight the most relevant skills, experience, and achievements from the resume that match the job description.
3. Use a professional and engaging tone.
4. Quantify achievements where possible (e.g., "improved performance by 20%").
5. Show enthusiasm for the role and alignment with the company’s goals.
6. Avoid generic statements — make it specific and impactful.
7. Keep it concise (around 250–400 words).

Output Format:
- Start with a professional greeting (e.g., "Dear Hiring Manager,").
- 1st paragraph: Strong introduction and role interest.
- 2nd–3rd paragraph: Relevant experience and achievements.
- Final paragraph: Closing statement with enthusiasm and call to action.
- End with a professional sign-off.

Important:
- Do NOT repeat the resume verbatim.
- Do NOT include placeholders like [Your Name].
- Ensure the letter reads naturally and is ready to send.
"""
```

```python
for m in llm.stream(template):
  print(m.content, end="", flush = True)
```
