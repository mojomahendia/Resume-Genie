```python
import os
from langchain_openai import ChatOpenAI
from langchain_community.document_loaders import PyMuPDFLoader

from dotenv import load_dotenv
load_dotenv()
```

```python
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")

llm=ChatOpenAI(
    model_name="gpt-3.5-turbo"
)
llm.invoke("who is lincon. answer in 1 line")
```

```python
loader=PyMuPDFLoader("/Users/mojo/Generative AI/Projects/Resume-Genie/Manoj Kumar-Gen AI.pdf")
resume = loader.load()
# resume[0].page_content

# clean view
context = "\n\n".join([d.page_content for d in resume])
```

```python
from langchain_core.messages import HumanMessage, SystemMessage, AIMessage

chat_history = []

system_message = SystemMessage(
    content=f"""
    You are a professional career coach and resume mentor.

    You help with:
    - Career Guidance
    - Resume Improvements
    - Interview Preparation
    - Job Search Strategy
    - Skill Gap Analysis

    Candidate Resume:
    {context} """
)
```

```python
while True:
    user_input = input("you: ")
    
    if user_input.lower() in ['exit','quit']:
        print("\n Goodbye!!")
        break
    chat_history.append(HumanMessage(contect=user_input))

    # building message list
    messages = [system_message] + chat_history
    
    print("\nCoach: ",end=" ",flush=True)

    response_text = ""
    
    for chunk in llm.stream(messages):
        print(chunk.content, end=' ', flush=True)
        response_text = response_text+chunk.content
    chat_history.append(AIMessage(content=response_text))
    print("\n")
```
