```python
job_description="""
We are seeking a talented AI Engineer with at least 2 years of experience building AI-driven solutions. The ideal candidate will design, develop, and deploy scalable machine learning and generative AI systems to solve real-world problems.

Key Responsibilities:

Build and deploy AI/ML models and pipelines
Develop and optimize LLM-based applications (RAG, prompt engineering, etc.)
Work with large datasets for training and evaluation
Collaborate with cross-functional teams to integrate AI solutions
Monitor model performance and improve accuracy over time

Required Skills:

2+ years of experience building AI or ML solutions
Strong proficiency in Python and ML frameworks (TensorFlow, PyTorch)
Experience with LLMs, NLP, or Generative AI
Knowledge of cloud platforms (AWS, Azure, or GCP)
Familiarity with data pipelines and APIs"""
```

```python
template="""
You are a senior recruiter and advanced ATS (Applicant Tracking System) evaluator with deep hiring experience.

Your task is to evaluate how well a candidate's resume matches a job description using real-world hiring logic.

--------------------------------------------------
INPUTS
--------------------------------------------------
Candidate Resume:
{resume}

Job Description:
{job_description}

--------------------------------------------------
STEP 1: EXTRACT STRUCTURED REQUIREMENTS
--------------------------------------------------
From the job description, identify:

- MUST_HAVE_SKILLS (critical, non-negotiable)
- NICE_TO_HAVE_SKILLS (good but optional)
- REQUIRED_EXPERIENCE (years + domain)
- REQUIRED_EDUCATION
- KEY_RESPONSIBILITIES

--------------------------------------------------
STEP 2: CANDIDATE ANALYSIS
--------------------------------------------------
From the resume, extract:

- Skills (technical + soft)
- Total years of experience
- Relevant experience (aligned with role)
- Education
- Key achievements (with metrics if available)

--------------------------------------------------
STEP 3: ATS KEYWORD MATCHING
--------------------------------------------------
- Compare resume keywords with MUST_HAVE and NICE_TO_HAVE skills
- Calculate:
  - Exact keyword matches
  - Partial/semantic matches
  - Missing critical keywords
- Penalize keyword stuffing or irrelevant keywords

--------------------------------------------------
STEP 4: SCORING RUBRICS (TOTAL = 100)
--------------------------------------------------

1. Must-Have Skills Match (30 points)
   - Strong penalty if critical skills are missing

2. Experience & Seniority Fit (20 points)
   - Match years + level (junior/mid/senior)
   - Domain relevance

3. Responsibilities Alignment (15 points)
   - Overlap with job responsibilities

4. ATS Keyword Optimization (10 points)
   - Keyword coverage + relevance

5. Achievements & Impact (10 points)
   - Quantified results, business impact

6. Education & Certifications (5 points)

7. Nice-to-Have Skills Bonus (10 points)

--------------------------------------------------
STEP 5: KNOCKOUT CHECK (CRITICAL)
--------------------------------------------------
- If multiple MUST_HAVE_SKILLS are missing OR
- Experience is significantly below requirement

→ Mark as: "High Risk Rejection"

--------------------------------------------------
STEP 6: FINAL DECISION LOGIC
--------------------------------------------------
Based on score:

- 85–100 → Strong Hire / Shortlist
- 70–84 → Consider / Interview
- 50–69 → Borderline
- <50 → Reject

--------------------------------------------------
OUTPUT FORMAT (STRICT JSON)
--------------------------------------------------

{{
  "overall_score": <0-100>,
  "decision": "<Strong Hire | Consider | Borderline | Reject>",
  "risk_flag": "<None | Low | Medium | High Risk Rejection>",

  "requirement_extraction": {{
    "must_have_skills": [],
    "nice_to_have_skills": [],
    "required_experience": "",
    "required_education": "",
    "key_responsibilities": []
  }},

  "candidate_summary": {{
    "skills": [],
    "years_of_experience": "",
    "relevant_experience": "",
    "education": "",
    "key_achievements": []
  }},

  "category_scores": {{
    "must_have_skills": {{
      "score": <0-30>,
      "reason": ""
    }},
    "experience_seniority": {{
      "score": <0-20>,
      "reason": ""
    }},
    "responsibility_alignment": {{
      "score": <0-15>,
      "reason": ""
    }},
    "ats_keywords": {{
      "score": <0-10>,
      "reason": ""
    }},
    "achievements_impact": {{
      "score": <0-10>,
      "reason": ""
    }},
    "education": {{
      "score": <0-5>,
      "reason": ""
    }},
    "nice_to_have_bonus": {{
      "score": <0-10>,
      "reason": ""
    }}
  }},

  "keyword_analysis": {{
    "matched_keywords": [],
    "missing_critical_keywords": [],
    "partial_matches": []
  }},

  "strengths": [],
  "gaps": [],
  "recommendations": [
    "Add missing must-have skills explicitly",
    "Improve keyword alignment with job description",
    "Quantify achievements with metrics",
    "Tailor experience bullets to responsibilities"
  ],

  "summary": "<Concise recruiter-style evaluation>"
}}

--------------------------------------------------
IMPORTANT RULES
--------------------------------------------------
- Ensure category scores sum exactly to overall_score
- Be strict and realistic (simulate real hiring filtering)
- Penalize missing must-have skills heavily
- Do NOT output anything outside JSON
- Avoid hallucination — only use given data
"""
```

```python
for m in llm.stream(template):
  print(m.content, end="", flush = True)
```
